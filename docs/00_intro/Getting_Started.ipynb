## Getting Help

If you get stuck, these are good places to look:

1. The Julia Discourse (https://discourse.julialang.org/) has a lot of examples, and a decent number of HPC-centric examples and solutions.
2. For faster feedback, try the Julia Slack (https://julialang.org/community/)

Getting involved with the community:

1. Monthly HPC Call (https://julialang.org/community/working-groups/). We are also working (slowly) on dedicated HPC pages: https://juliaparallel.org/
2. BoFs at SC, JuliaCon (https://juliacon.org/2023/)

## How to get Julia:

Fast (but often outdated):

1. Debian/Ubuntu: `apt-get install julia`
2. macOS: `brew install --cask julia`

Bleeding edge:

3. Juliaup: https://github.com/JuliaLang/juliaup (Juliaup also provides as version multi-plexer)
4. Manual install: https://julialang.org/downloads/
5. Spack: https://spack.io/ (but often requires expert-level knowledge)

## Package Management

Two ways to manage packages:
1. You can use the `Pkg` module from within Julia.
2. From the Julia REPL you can type `]` to enter the Package Manager's CLI

Eg. to add the `CSV` package using the `Pkg` module run:

```julia
import Pkg
Pkg.add("CSV")
```

````julia
import Pkg
Pkg.rm("CSV")
````

````julia
import Pkg
Pkg.add("CSV")
````

When you're in the REPL you can type `]` to enter the Package Manager's CLI (note that the promt turns blue and has the format `(environment name) pkg>`). Eg. to add the `CSV` package using the Package Manager's CLI in the REPL, type: `add CSV` in the Package manager

![repl](pkg.png)

(hit ESC to go back to normal mode)

````julia
cd(mktempdir())
pwd() # print current directory
````

````julia
Pkg.activate(".")
````

````julia
Pkg.add("CSV")
````
